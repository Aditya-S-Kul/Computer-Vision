# 1 Convolutional Blocks of ResNet18

##### Note: All the code and training part of entire question is done in below cell. Even losses and some other things are printed. All the other things such as plots,etc. is reported in markdown cells followed by this cell. 

```python
# q1.ipynb

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F

import wandb
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, confusion_matrix
import numpy as np



device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print("Using device:", device)


# 1. Load the dataset from .pt files 
# Convert images to float and scale them to [0,1]
train_data = torch.load("train_data.pt").float() / 255.0
train_labels = torch.load("train_labels.pt")
test_data = torch.load("test_data.pt").float() / 255.0
test_labels = torch.load("test_labels.pt")


# 2. Define a custom Dataset class with an optional transform.
class CustomDataset(Dataset):
    def __init__(self, data, labels, transform=None):

        # data: a tensor of images, assumed to be of shape (N, C, H, W)
        self.data = data
        # labels: a tensor of labels
        self.labels = labels
        # transform: a function to transform the image (e.g. resize)
        self.transform = transform

        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        image = self.data[idx]   # tensor image (now float, scaled to [0,1])
        label = self.labels[idx]
        if self.transform:
            image = self.transform(image)
        return image, label



def evaluate_model(model, data_loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in data_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    avg_loss = running_loss / total
    accuracy = correct / total
    f1 = f1_score(all_labels, all_preds, average='weighted')
    conf_matrix = confusion_matrix(all_labels, all_preds)
    return avg_loss, accuracy, f1, conf_matrix

def train_model(model, train_loader, test_loader, optimizer, criterion, num_epochs, device, run_name="experiment"):
    model.to(device)
    wandb.init(project="ResNet_Experiments", name=run_name, reinit=True)
    
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
        epoch_loss = running_loss / total
        epoch_acc = correct / total
        
        # Evaluate on the test set
        test_loss, test_acc, test_f1, conf_matrix = evaluate_model(model, test_loader, criterion, device)
        print(f"[{run_name}] Epoch {epoch+1}/{num_epochs} - Train Loss: {epoch_loss:.4f}, Train Acc: {epoch_acc:.4f}, "
              f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}, Test F1: {test_f1:.4f}")
        
        # Log metrics to wandb
        wandb.log({
            "epoch": epoch+1, 
            "train_loss": epoch_loss, 
            "train_accuracy": epoch_acc,
            "test_loss": test_loss, 
            "test_accuracy": test_acc,
            "test_f1": test_f1
        })
    return model


def modify_final_layer(model, num_classes):
    # Modify the final fully connected layer to have the correct number of classes.
    in_features = model.fc.in_features
    model.fc = nn.Linear(in_features, num_classes)
    return model

def modify_resnet(model, modification):
# In this function only conv1 is replaced and reinitialized; all other layers retain their existing weights.

    # "A": Change conv1 to kernel_size=3(no aggressive downsampling), stride=1, padding=1 and remove maxpool.
    if modification == "A":
        in_channels = model.conv1.in_channels
        out_channels = model.conv1.out_channels
        model.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, stride=1, padding=1, bias=False)
        # Reinitialize conv1 (whether pretrained or not, the new layer has random weights)
        nn.init.kaiming_normal_(model.conv1.weight, mode='fan_out', nonlinearity='relu')
        model.maxpool = nn.Identity()  # remove max pooling

    # "B": Keep conv1 unchanged but remove maxpool to avoid losing resolution.
    elif modification == "B":
        model.maxpool = nn.Identity()

    # "C": Change conv1 to kernel_size=5(slightly larger kernel), stride=1, padding=2 and replace maxpool with a smaller one.
    elif modification == "C":
        in_channels = model.conv1.in_channels
        out_channels = model.conv1.out_channels
        model.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=5, stride=1, padding=2, bias=False)
        nn.init.kaiming_normal_(model.conv1.weight, mode='fan_out', nonlinearity='relu')
        model.maxpool = nn.MaxPool2d(kernel_size=2, stride=1)
    else:
        raise ValueError("Modification must be one of 'A', 'B', or 'C'.")
    
    return model



def run_experiment(exp_name, image_size, model_source="scratch", modification=None, num_epochs=5, batch_size=64):
   
    # Initialize wandb (each experiment runs a new wandb run)
    wandb.init(project="ResNet_Experiments", name=exp_name, reinit=True)
    
    # Define a transform function.
    # Note: our images are already tensors. For resizing, we use torch.nn.functional.interpolate.
    if image_size is None:
        transform = None
    else:
        transform = lambda x: F.interpolate(x.unsqueeze(0), size=image_size, mode='bilinear', align_corners=False).squeeze(0)
    
    # Create datasets and dataloaders.
    train_dataset = CustomDataset(train_data, train_labels, transform=transform)
    test_dataset = CustomDataset(test_data, test_labels, transform=transform)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader  = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    

    num_classes = len(torch.unique(train_labels))
    
    # Load the standard ResNet18 model.
    if model_source == "scratch":
        model = torchvision.models.resnet18(pretrained=False)
    elif model_source == "pretrained":
        model = torchvision.models.resnet18(pretrained=True)
    else:
        raise ValueError("model_source must be 'scratch' or 'pretrained'")
    

    model = modify_final_layer(model, num_classes)
    

    if modification is not None:
        model = modify_resnet(model, modification, pretrained=(model_source=="pretrained"))
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    trained_model = train_model(model, train_loader, test_loader, optimizer, criterion, num_epochs, device, run_name=exp_name)
    
    # Evaluate final performance.
    test_loss, test_acc, test_f1, conf_matrix = evaluate_model(trained_model, test_loader, criterion, device)
    print(f"[{exp_name}] Final Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.4f}, Test F1 Score: {test_f1:.4f}")
    wandb.log({
        "final_test_loss": test_loss,
        "final_test_accuracy": test_acc,
        "final_test_f1": test_f1
    })
    wandb.finish()
    
    # Plot confusion matrix.
    plt.figure(figsize=(6,6))
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix: {exp_name}')
    plt.colorbar()
    tick_marks = np.arange(num_classes)
    plt.xticks(tick_marks)
    plt.yticks(tick_marks)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()



# Experiment 1.1: Baseline on original 36x36 images
# a) ResNet18 from scratch
run_experiment(exp_name="Baseline_Scratch_36", image_size=None, model_source="scratch", modification=None, num_epochs=5, batch_size=64)

# b) Pretrained ResNet18 (modify final layer for number of classes)
run_experiment(exp_name="Baseline_Pretrained_36", image_size=None, model_source="pretrained", modification=None, num_epochs=5, batch_size=64)

# Experiment 1.2: Training on resized images 224x224
# a) From scratch
run_experiment(exp_name="Baseline_Scratch_224", image_size=(224,224), model_source="scratch", modification=None, num_epochs=5, batch_size=64)

# b) Pretrained
run_experiment(exp_name="Baseline_Pretrained_224", image_size=(224,224), model_source="pretrained", modification=None, num_epochs=5, batch_size=64)



# Experiment 1.3: Modified architectures to better suit 36x36 images.
# three modifications: "A", "B", "C" (defined above).

# (i) Training from scratch with modifications.
for mod in ["A", "B", "C"]:
    run_experiment(exp_name=f"Modified_{mod}_Scratch_36", image_size=None, model_source="scratch", modification=mod, num_epochs=5, batch_size=64)

# (ii) Pretrained models with modifications.
for mod in ["A", "B", "C"]:
    run_experiment(exp_name=f"Modified_{mod}_Pretrained_36", image_size=None, model_source="pretrained", modification=mod, num_epochs=5, batch_size=64)
```

### 4. Spatial dimensions:  

For a 36x36 input image using the standard ResNet18:  
- After conv1 (7x7, stride=2, padding=3): 64 x 18 x 18
- After maxpool (3x3, stride=2, padding=1): 64 x 9 x 9
- After layer1: 64 x 9 x 9
- After layer2: approx. 128 x 5 x 5
- After layer3: approx. 256 x 3 x 3
- After layer4: approx. 512 x 2 x 2  
So, just before average pooling the spatial dimensions are 2 x 2.

## 1.4 Comparison 

### 1. Comparison of all trained models  

In terms of accuracy, pretrained models generally performed better than from scratch models. In the baseline case (36×36 input), the pretrained model achieved 0.88 accuracy, whereas the from scratch model reached 0.80. This performance boost in the pretrained case is because of transfer learning, where the model benefits from prior knowledge, especially since we haven't changed the architecture except for the final layer. Since the task is similar to the original dataset, transfer learning is effective.  

However, when we increased the image size to 224×224, we saw an even better accuracy, mainly because this resolution matches the model’s original architecture, meaning the model retains most of its pre-learned feature extraction abilities. But, this came at a huge cost in terms of training speed. The training time increased 8-9 times (from 2.5-3 minutes to 22-24 minutes), making it much slower than the 36×36 setup.  

When we introduced modifications to the model, an interesting trend was observed. From scratch models outperformed pretrained models in some cases, which is opposite to the usual expectation. This unexpected behavior was mainly due to the way initialization was handled. In the pretrained models, the first layer was randomly initialized, while the deeper layers retained the pre-learned weights. Since the first layer learns basic features like edges and textures, resetting it while keeping the rest of the model unchanged caused problems. The pretrained model had to relearn both fundamental features and high-level dataset-specific features simultaneously, which made training difficult. On the other hand, from scratch models had no such issue. Even though they started with random weights, they learned everything from the dataset itself in a structured way, allowing them to perform better in some cases.  

Another aspect that affected performance was the kernel size in the first convolutional layer. When we changed the kernel size from 7×7 (original ResNet18) to 3×3 or 5×5, the accuracy changed depending on the setup. A smaller kernel (3×3) helped retain more local details but slightly reduced the model’s ability to generalize. The 5×5 kernel, on the other hand, acted as a middle ground—capturing slightly larger patterns without aggressive downsampling. Removing the max pooling layer also had an impact, as it preserved more spatial information, which was useful for small input sizes like 36×36.  

In summary, pretrained models generally provided better accuracy, but from scratch models showed better performance when modifications were made. Increasing image size improved accuracy but significantly slowed down training. The way initialization was handled had a major impact, particularly in pretrained models, where resetting the first layer negatively affected learning. Kernel size and pooling operations also influenced performance, with smaller kernels helping retain local details and modified pooling strategies improving spatial retention for small images.

### 1.4.3

The differences in performance come from transfer learning, input size, architecture changes, and initialization. Pretrained models usually perform better because they already know useful features, but if we change the first layer, they struggle since they lose important early features. From scratch models don’t have this problem because they learn everything from our dataset. Input size also plays a role—bigger images (224×224) match ResNet18’s original design, giving better accuracy, but training takes much longer compared to smaller images. Architecture changes like modifying the kernel size or removing max pooling affect how much detail the model keeps; smaller kernels preserve fine details, while bigger ones capture broader patterns. Initialization also impacts performance—pretrained models suffer when the first layer is reset because they forget basic features while later layers still have high-level ones, making learning harder. From scratch models don’t face this issue since they start fresh and learn everything together. Overall, these differences arise based on how well the model adapts, whether it can use past knowledge, and how changes in design affect learning.
